# <b>5<span style='color:#0000FF'>|</span> Data Visualization </b>

```python
df.columns
```

```python
df.drop(index=0,inplace=True)
```

```python
def Bar(feature):
    global df
    fig,ax = plt.subplots(figsize = (15,10))
    TOP_10=df[['Country,\nOther',feature]].groupby(['Country,\nOther']).sum().sort_values(feature,axis=0,ascending=False)[:10]
    TOP_10.plot(kind='bar',alpha=0.7,ax=ax)
    plt.xticks(rotation = 50)
    plt.xlabel("Country",fontsize=15)
    plt.ylabel(feature,fontsize=15)
    plt.title("Top 10 "+feature,fontsize=20)
    plt.set_cmap('summer')
    plt.show()
    return TOP_10
```

```python
Bar('Total\nDeaths')
```

```python
Bar('Total\nCases')
```

```python
TOP_10=df[['Country,\nOther','Total\nCases','Total\nDeaths']].groupby(['Country,\nOther']).sum().sort_values('Total\nDeaths',axis=0,ascending=False)[:10]
```

```python
TOP_10
```

```python
TOP_10.plot.pie(subplots=True, 
                    autopct='%1.1f%%',
                    figsize=(15, 10),
                    startangle=90,     # start angle 90° (Africa)
                    shadow=True,
                    labels = TOP_10.index)
```
