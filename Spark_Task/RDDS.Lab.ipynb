### Create RDD and apply some functions ...

```
nums = sc.parallelize([1, 2, 3,5,12,20])
```

1. Square nums
2. find even nums
3. find summation of nums

```
squares = nums.map(lambda x: x*x)
even = squares.filter(lambda x: x % 2 == 0) 
merged = nums.reduce(lambda x, y: x + y) 
```

```
print(squares.collect())
```

```
print(even.collect())
```

```
print(merged)
```
